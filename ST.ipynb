## Przypomnienie wiadomości o funkcjach - jak tworzyć docstring

```python
def silnia(n):
  """
  Funkcja liczy silnię.

  Args:
    n (int): Liczba, której wyznaczymy silnię

  Return:
    (int): wartośc silnii dla podanego n

  Raises:
    TypeError:
      Gdy podamy wartość, która nie jest typu int.
    ValueError:
      Gdy podamy wartość ujemną.

  Examples:
  >>> silnia(3)
  6

  """
  if not type(n) == int:
    raise TypeError("Oczekuje wartosci int.")
  if n<0:
    raise ValueError("Oczekuje wartosci nieujemnych.")
  if n == 0 or n==1:
    return 1
  else:
    return silnia(n-1)*n
```

```python
try:
  for i in range(9):
    print(f"{i}! = {silnia(i)}")
  n = input ("Podaj wartość policzę silnię  ")
  print(f"Wartość {n}! = {silnia(n)}")
except Exception as wyj:
  print("Uwaga błąd: "+str(wyj))
```

```python
def pad_iter(n):
    p0 = 1
    p1 = 1
    p2 = 1
    for i in range(3, n+1):
        pn = p0 + p1
        p0 = p1
        p1 = p2
        p2 = pn
    return p2
```
